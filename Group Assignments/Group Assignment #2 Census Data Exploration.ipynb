### Importing Second Dataset and Preliminary Inspection

Purpose: I use geopandas to access and read the geoson file that I retrieved from Census Reporter. This particular dataset reveals block groups of the LA that are filtered by Median Household Income. I then use the `info()`, `head()`, `plot()` command to get a cursory glance of the size and components of the this dataset (e.g., number of rows and columns, data types, etc.).

```python
bg=gpd.read_file('Data/acs2019_5yr_B19013_15000US060372318003.geojson')
```

```python
bg.info()
```

```python
bg.head()
```

```python
bg.plot()
```

### Trimming the Data

Purpose: I trim the `bg` dataset to only columns that interest me: `geoid`, `name`, `B19013001 (Median Household Income)`, and `geometry`. I then update `bg` dataset and `copy()`.

```python
desired_columns=['geoid','name','B19013001','geometry']

bg[desired_columns]
```

```python
bg=bg[desired_columns].copy()
```

### Renaming Column Headers

Purpose: I rename obscure headers in the dataset to something more appopriate: 'B19013001' to 'MHI.'

```python
bg.columns = ['geoid','name','MHI','geometry']
bg
```

### Query the Dataset and Plotting

Purpose: I further query the dataframe by filtering Block Groups that have a MHI of less than 40000. I assign the subset to a new variable and generate a map that reflect the query.

```python
bg[bg['MHI'] < 40000.0]
```

```python
low_income=bg[bg['MHI'] < 40000.0] #low income block groups under $40k
low_income.plot(figsize=(15,13),
                 column='MHI',
                 legend=True, 
                 scheme='NaturalBreaks')
```

## Spatial Join and Preliminary Inspection of Koreatown

Purpose: I spatial join the `low income` and `ktown` dataframes using `sjoin()`. This merges the geometry objects based on their spatial relationship to one another and outputs a subset where the low income block groups touch Koreatown. To prove that its a new geodataframe, I use `head` to look at the size of the dataframe.

```python
ktown_bg = gpd.sjoin(low_income, ktown, how="inner", op="intersects")
type(ktown_bg)
```

```python
ktown_bg.head()
```

### Simple Stats and Plot

Purpose: I use `describe()` to get a statistical breakdown of the 'MHI' column of the `ktown_bg` dataframe. I then generate a histogram to contextualize the statistical information. Additionaly, I generate a map of Koreatown that are distinguised by different income levels.

```python
print(ktown_bg["MHI"].describe())
print(ktown_bg["MHI"].plot.hist(bins=3))
```

```python
ktown_bg.plot(figsize=(12,10),
                 column='MHI',
                 legend=True, 
                 scheme='quantiles')
```

### Adding a Base Map

Purpose: I reproject the geodataframe to Web Mercator and generate a multi-layered map of Koreatown.

```python
ktown_bg_web_mercator = ktown_bg.to_crs(epsg=3857) 
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
ktown_bg_web_mercator[ktown_bg_web_mercator['MHI']>0].plot(ax=ax, alpha=0.8, legend= True, column= "MHI")
ax.axis('off')
ax.set_title('Block Groups with Median Household Income less than $40,000 in Koreatown',fontsize=16)
ctx.add_basemap(ax)
```
